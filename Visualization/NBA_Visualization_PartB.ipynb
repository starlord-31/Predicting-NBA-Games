```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data=pd.read_csv('nba_games.csv')
data.head()
```

```python
data.columns[60:80]
```

```python
data = data.drop("mp", axis = 1)
data = data.drop("mp.1", axis = 1)
data = data.drop("Unnamed: 0", axis = 1)
data = data.replace({True: 1, False: 0})
```

```python
X = data.drop("won", axis = 1)
y = data['won']
```

```python
y = y.replace({True: 1, False: 0})
```

```python
plt.title("Games Won by 3 Point %")
sns.lineplot(data=data,x='3p%',y='won')
```

```python
plt.title("Games Won by Oppenents Field Goal %")
sns.lineplot(data=data,x='fg%_opp',y='won')
plt.show()
```

```python
grouped = data.groupby('team').mean().reset_index()
```

```python
grouped
```

```python
grouped['won'].nlargest(1)
```

```python
grouped['won'].nsmallest(1)
```

```python
data2 = data[(data['team'].isin(['GSW','ORL']))]
```

```python
data2.columns[:50]
```

```python
data2.columns[50:]
```

```python
fig, axes = plt.subplots(1, 3,figsize=(15,5))
axes[0].set_title("Field Goal % ORL vs GSW")
axes[1].set_title("Steals ORL vs GSW")
axes[2].set_title("Turnovers ORL vs GSW")
sns.kdeplot(data=data2,x='fg%',hue='team',ax=axes[0])
sns.kdeplot(data=data2,x='stl',hue='team',ax=axes[1])
sns.kdeplot(data=data2,x='tov',hue='team',ax=axes[2])
```

```python
fig, axes = plt.subplots(1, 2,figsize=(15,5))
axes[0].set_title("Points Scored ORL vs GSW")
axes[1].set_title("Points Scored Against ORL vs GSW")
sns.violinplot(data=data2, x='team', y='pts',ax=axes[0])
sns.violinplot(data=data2, x='team', y='pts_opp',ax=axes[1])
plt.show()
```

```python
defense = grouped[(grouped['team'].isin(['SAS','MIA','PHI','OKC','IND']))]
offense = grouped[(grouped['team'].isin(['DEN','LAC','HOU','POR','MIL']))]
```

```python
fig, axes = plt.subplots(1, 2,figsize=(15,5))
axes[0].set_title("Defensive Teams Win %")
axes[1].set_title("Offensive Teams Win %")
sns.barplot(data=defense,x='team',y='won',ax=axes[0])
sns.barplot(data=offense,x='team',y='won',ax=axes[1])
print("Mean win percentage for defensive teams:",defense['won'].mean())
print("Mean win percentage for offensive teams:",offense['won'].mean())
```
