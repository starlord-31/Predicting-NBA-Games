```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('nba_games.csv')
df = df.sort_values(by='date', ascending=True)
df = df.reset_index(drop=True)
df.head()
```

```python
df = df.drop(['Unnamed: 0', 'mp', 'mp.1', 'mp_opp', 'mp_opp.1', '+/-', '+/-_opp',  'mp_max', 'mp_max.1', 'mp_max_opp', 'mp_max_opp.1'], axis=1)
cols_to_front = ['date', 'team', 'total', 'team_opp', 'total_opp', 'won']

df = df[cols_to_front + [col for col in df.columns if col not in cols_to_front]]

df.head()
```

```python
df.shape
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['season'].value_counts()
```

```python
relevant_cols = ['fg', 'fga', 'fg%', '3p', '3pa', '3p%', 'ft', 'fta', 'ft%', 'orb', 'drb', 'trb', 'ast', 'stl', 'blk', 'tov', 'pf', 'pts']
avg_cols = ['season_avg_' + col for col in relevant_cols]
df[avg_cols] = 0
df[avg_cols].head() 
```

```python
df_2016 = df[df['season'] == 2016].copy()
df_2016 = df_2016.reset_index(drop=True)

df_2017 = df[df['season'] == 2017].copy()
df_2017 = df_2017.reset_index(drop=True)

df_2018 = df[df['season'] == 2018].copy()
df_2018 = df_2018.reset_index(drop=True)

df_2019 = df[df['season'] == 2019].copy()
df_2019 = df_2019.reset_index(drop=True)

df_2020 = df[df['season'] == 2020].copy()
df_2020 = df_2020.reset_index(drop=True)

df_2021 = df[df['season'] == 2021].copy()
df_2021 = df_2021.reset_index(drop=True)

df_2022 = df[df['season'] == 2022].copy()
df_2022 = df_2022.reset_index(drop=True)

seasonal_dfs = [df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022]
```

```python
teams = df['team'].unique()
team_frames_2016 = {team: df_2016[df_2016['team'] == team].copy() for team in teams}
team_frames_2017 = {team: df_2017[df_2017['team'] == team].copy() for team in teams}
team_frames_2018 = {team: df_2018[df_2018['team'] == team].copy() for team in teams}
team_frames_2019 = {team: df_2019[df_2019['team'] == team].copy() for team in teams}
team_frames_2020 = {team: df_2020[df_2020['team'] == team].copy() for team in teams}
team_frames_2021 = {team: df_2021[df_2021['team'] == team].copy() for team in teams}
team_frames_2022 = {team: df_2022[df_2022['team'] == team].copy() for team in teams}

seasonal_team_frames = [team_frames_2016, team_frames_2017, team_frames_2018, team_frames_2019, team_frames_2020, team_frames_2021, team_frames_2022]

```

```python
def calculate_season_avgs(df, stats):
    df_expnaded = df[stats].expanding().mean()
    return df_expnaded

for team_frames in seasonal_team_frames:
    for team, team_df in team_frames.items():
        team_df[avg_cols] = calculate_season_avgs(team_df, relevant_cols)
```

```python
df_2016 = pd.concat([team_df for team_df in team_frames_2016.values()])
df_2016 = df_2016.sort_values(by='date', ascending=True)

df_2017 = pd.concat([team_df for team_df in team_frames_2017.values()])
df_2017 = df_2017.sort_values(by='date', ascending=True)

df_2018 = pd.concat([team_df for team_df in team_frames_2018.values()])
df_2018 = df_2018.sort_values(by='date', ascending=True)

df_2019 = pd.concat([team_df for team_df in team_frames_2019.values()])
df_2019 = df_2019.sort_values(by='date', ascending=True)

df_2020 = pd.concat([team_df for team_df in team_frames_2020.values()])
df_2020 = df_2020.sort_values(by='date', ascending=True)

df_2021 = pd.concat([team_df for team_df in team_frames_2021.values()])
df_2021 = df_2021.sort_values(by='date', ascending=True)

df_2022 = pd.concat([team_df for team_df in team_frames_2022.values()])
df_2022 = df_2022.sort_values(by='date', ascending=True)

seasonal_dfs = [df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022]
df = pd.concat(seasonal_dfs)
```

```python
df = df.sort_values(by='date', ascending=True)
df = df.reset_index(drop=True)

df.head(20)
```

```python
df.to_csv('nba_games_cleaned_with_avgs.csv')
```
